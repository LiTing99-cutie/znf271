```python
import pandas as pd 
import pybedtools
from pybedtools import BedTool
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import numpy as np
import argparse
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv("/home/user/data2/lit/project/ZNF271/02-APA-1/evolution-diff/r.terminal_exon.genePredExt",sep='\t',header=None)
data.columns=["chr","strand","start","end","gene_id"]
data.head()
data_reverse=data[data.strand=='-']
data_forward=data[data.strand=='+']
data_reverse.head()
```

```python
tmp = data_forward[data_forward.gene_id=="ZNF271"]
# tmp = data_forward

i=1
col = 'start'

# id_start
tmp["id_start"]=tmp["gene_id"]+"_"+tmp[col].apply(str)
# count id_start
id_start_c=tmp["id_start"].value_counts().reset_index()
id_start_c.columns=['id_start','c']
id=id_start_c.id_start.str.split("_",expand=True).iloc[:,0]
id_start_c['id']=id
# max count
id_start_c_max=id_start_c.groupby(["id"]).apply(lambda t:t[t.c==t.c.max()])
id_start_c_max
```

```python
AbFirstOrNot=[]
for id in id_start_c_max['id'].unique():
    if len(id_start_c_max[id_start_c_max.id==id]) ==1:
            AbFirstOrNot.append(id_start_c_max[id_start_c_max.id==id]['id_start'].values)
AbFirstOrNot=pd.DataFrame(AbFirstOrNot)
AbFirstOrNot.columns=['id_start']
AbFirstOrNot
```

```python
len(id_start_c_max[id_start_c_max.id=='ZNF271'])
```

```python
final=pd.merge(tmp,AbFirstOrNot,on="id_start",how="inner").iloc[:,:5]
```
